## Основные теоретические положения
Случайный вектор $ \overline\xi = (\xi_1,\, ...,\, \xi_n)^{\text{T}}\, :\, \Omega \rightarrow \mathbb{R}^n $ - называют нормальным (Гауссовским), если существует вектор независимых стандартных нормальных случайных величин $\overline{\eta} = (\eta_1,\, ...,\, \eta_m)^{\text{T}},$ матрица $A$ размерности $ m \times n $ и вещественный вектор $ B = (b_1,\, ...\, b_n)^\text{T}, $ такие что:
$$
    \overline{\xi} = A\overline{\eta} + B.
$$
Причем $ \overline\xi \sim \mathcal{N}(\mu, \Sigma), $ где матожидание $ \mu = \mathbb{E}\overline\xi = (E\xi_1,\, ...,\, E\xi_n)^{\text{T}}, $ ковариационная матрица $ \Sigma = \{\sigma_{ij}\}_{i,j = 1}^{n},\, \sigma_{ij} = \mathop{\mathrm{cov}}(\xi_i,\, \xi_j), $ а плотность распределения имеет вид:
$$
    \begin{equation}
        \rho_{\overline{\xi}} = \frac{1}{(2\pi)^{\frac{n}{2}}\sqrt{\det{\Sigma}}}\exp\left ( -\frac{1}{2}\left(\overline{x}-\mu\right)^\text{T}\Sigma^{-1}\left(\overline{x}-\mu\right)\right ), \\
        \overline{x} = (x_1,\, ...,\, x_n)^\text{T}.
    \end{equation}
$$
### Свойства
1. Линейная комбинация случайных переменных, принадлежащих многомерному нормальному распределению, также является нормально распределенной.
2. Условное и безусловное распределения являются нормальным распределением.
3. Параметры многомерного нормального распределения полностью описывают его форму.
4. Многомерное нормальное распределение устойчиво относительно линейных преобразований, т.е., если $ \overline{\xi} \sim \mathcal{N}(\mu, \Sigma), $ матрица $A$ размера $ m \times n $ - произвольная, $B$ - произвольный вектор размера $m \times 1,$ то
$$
    A\overline{\xi} + B \sim \mathcal{N}(A\mu + B, A\Sigma A^{\text{T}}).
$$

#### Задание №5
Найдем условное распределение $\xi | \eta,$ вспомним, что $\mathbb{\eta} = -1,$ $D_\eta = \frac{1}{3},$ тогда:
$$
\rho_\eta(y) = \frac{1}{\sqrt{\frac{2\pi}{3}}}\exp\left (-\frac{1}{2}\frac{(y+1)^2}{\left (\frac{1}{3}\right )}\right ).
$$
Плотность условного распределения равна
$$
\rho_{\xi | \eta = y}(x) = \frac{\rho_{\xi, \eta}(x, y)}{\rho_\eta(y)} = 
\frac
{\frac{\sqrt{6}}{2\pi}\exp \left ( -\frac{1}{2}(2x^2 + 5y^2 + 6y + 4xy + 3)\right )}
{\frac{1}{\sqrt{\frac{2\pi}{3}}}\exp \left ( -\frac{1}{2}(3y^2 + 6y + 3)\right)} = \\
\frac{1}{\sqrt{\pi}}\exp \left ( -\frac{1}{2} (2x^2+5y^2 + 6y + 4xy + 3 - (3y^2 + 6y +3))\right) = \\
\frac{1}{\sqrt{\pi}}\exp \left ( -\frac{1}{2} (2(x^2 + 2xy + y^2))\right) =
\frac{1}{\sqrt{2\pi}\cdot \sqrt{\frac{1}{2}}}\exp \left ( -\frac{1}{2} \cdot \frac{(x+y)^2}{\left(\frac{1}{2}\right)}\right);
$$
Таким образом:
$$
\mathbb{E}(\xi|\eta=y) = -y,\\
\mathbb{D}(\xi | \eta=y) = \frac{1}{2}.
$$
